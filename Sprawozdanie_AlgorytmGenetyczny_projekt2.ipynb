## 1. Cel Projektu
Celem projektu było rozszerzenie klasycznego algorytmu genetycznego (z Projektu 1) o reprezentację rzeczywistą chromosomu, a następnie przeprowadzenie wnikliwej analizy porównawczej obu podejść. Zaimplementowano dedykowane operatory genetyczne dla liczb zmiennoprzecinkowych:
* **Krzyżowania:** Arytmetyczne, Linearne, Uśredniające, BLX-α, BLX-α-β.
* **Mutacje:** Równomierna, Gaussa.

## 2. Technologie i Wymagania środowiskowe
Projekt opiera się na tym samym stabilnym stosie technologicznym co Projekt 1:
* **Język:** Python 3.10+
* **Biblioteki:** `numpy` (obliczenia wektorowe), `customtkinter` (interfejs graficzny), `matplotlib` (wizualizacja zbieżności), `pandas` (analiza logów i wyników).

Aby uruchomić kod oraz wygenerować niniejsze sprawozdanie, należy zainstalować wymagane pakiety:

`pip install numpy customtkinter matplotlib pandas notebook`
