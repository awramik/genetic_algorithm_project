## 3. Opis wybranych funkcji testowych i ich optima

W badaniach nad skutecznością zaimplementowanego Algorytmu Genetycznego wykorzystano funkcję **Hypersphere**. Jest to klasyczna funkcja testowa, charakteryzująca się unimodalnością, co pozwala na precyzyjną ocenę zdolności algorytmu do lokalizacji minimum globalnego bez ryzyka utknięcia w minimach lokalnych.

### Definicja matematyczna:
Funkcja ta jest zdefiniowana jako suma kwadratów zmiennych decyzyjnych $x_i$:

$$f(\mathbf{x}) = \sum_{i=1}^{n} x_i^2$$

### Parametry badania:
* **Wymiarowość (N):** Badania przeprowadzono dla $N \in \{2, 10, 27\}$.
* **Dziedzina poszukiwań:** Każda zmienna $x_i$ ograniczona jest przedziałem $[-5, 5]$.
* **Charakterystyka:** Funkcja jest ciągła, wypukła i różniczkowalna.

### Optimum globalne:
Algorytm dąży do znalezienia wektora wartości, dla którego funkcja przyjmuje wartość minimalną:
* **Punkt optimum:** $\mathbf{x}^* = (0, 0, \dots, 0)$
* **Wartość optimum:** $f(\mathbf{x}^*) = 0$

W przypadku testów maksymalizacji, ze względu na symetrię dziedziny $[-5, 5]$, punkty optymalne znajdują się na krawędziach przestrzeni poszukiwań (np. wektory składające się z samych piątek), gdzie wartość funkcji wynosi:
$$f(\mathbf{x}_{max}) = n \cdot 5^2$$
Dla $N=5$, maksimum wynosi dokładnie **125.0**, co zostało potwierdzone podczas testów metod krzyżowania.

## Importy i silnik 

````python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

def get_params(txt_path):
    params = {}
    if os.path.exists(txt_path):
        with open(txt_path, 'r', encoding='utf-8') as f:
            for line in f:
                if ':' in line:
                    key, val = line.split(':', 1)
                    params[key.strip()] = val.strip()
    return params

def load_all_experiments():
    files = sorted(glob.glob("results/logs/*_history.csv"))
    return [{'p': get_params(f.replace("_history.csv", "_params.txt")), 
             'd': pd.read_csv(f), 'n': os.path.basename(f)} for f in files]

all_exp = load_all_experiments()
print(f"Pomyślnie załadowano {len(all_exp)} eksperymentów.")
````

## Zbiorcze Porównanie Eksperymentów

W tej części skupiamy się tylko na zestawieniu linii "Best" ze wszystkich plików, aby zobaczyć tempo zbieżności różnych konfiguracji.


````python
# --- WYKRES 1: PEŁNA SKALA ---
plt.figure(figsize=(12, 6), facecolor='#F0F4F9')
ax1 = plt.gca()
ax1.set_facecolor('#FFFFFF')

for e in all_exp:
    label = f"{e['p'].get('EKSPERYMENT', e['n'])}"
    plt.plot(e['d']['epoch'], e['d']['best_value'], linewidth=1.5, label=label)

plt.title("Zestawienie Najlepszych Wyników", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Epoka"); plt.ylabel("Wartość Funkcji")
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8, ncol=1)
plt.tight_layout()
plt.show()

# --- WYKRES 2: ZBLIŻENIE ---
plt.figure(figsize=(12, 7), facecolor='#F0F4F9')
ax2 = plt.gca()
ax2.set_facecolor('#FFFFFF')

for e in all_exp:
    label = f"{e['p'].get('EKSPERYMENT', e['n'])}"
    plt.plot(e['d']['epoch'], e['d']['best_value'], linewidth=2.5, label=label, alpha=0.8)

plt.title("Zestawienie Najlepszych Wyników - Zbliżenie ", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Epoka", fontsize=11); plt.ylabel("Wartość Funkcji", fontsize=11)

# MAGIA PRZYCIĘCIA SIATKI:
plt.ylim(-0.5, 10) 

plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
plt.tight_layout()
plt.show()
````


## Szczegółowe Raporty Indywidualne


````python
print("\n" + "═"*70)
print("  INDYWIDUALNE RAPORTY TECHNICZNE")
print("═"*70 + "\n")

for e in all_exp:
    # Wykres Best vs Average
    plt.figure(figsize=(10, 4), facecolor='#F8FAFC')
    plt.plot(e['d']['epoch'], e['d']['best_value'], color='#2563EB', label='Best', linewidth=2)
    plt.plot(e['d']['epoch'], e['d']['average_value'], color='#F59E0B', linestyle='--', label='Average', alpha=0.7)
    
    plt.title(f"Eksperyment: {e['p'].get('EKSPERYMENT', e['n'])}", fontsize=11, fontweight='bold')
    plt.grid(True, alpha=0.3); plt.legend(); plt.show()
    
    # Tabela parametrów
    print("PARAMETRY KONFIGURACYJNE:")
    for k, v in e['p'].items():
        if v and not k.startswith("==="):
            print(f"  • {k:.<35} {v}")
    print("\n" + "─"*85 + "\n")
````



## Tabela Wyników


````python
summary_data = []
for e in all_exp:
    summary_data.append({
        "Eksperyment": e['p'].get('EKSPERYMENT'),
        "Najlepszy Wynik": e['d']['best_value'].iloc[-1],
        "Średni Wynik": e['d']['average_value'].iloc[-1],
        "Czas [s]": e['p'].get('Czas wykonania algorytmu')
    })

df_summary = pd.DataFrame(summary_data)
display(df_summary)
````

W ramach projektu przeprowadziłam serię eksperymentów mających na celu zbadanie wpływu poszczególnych parametrów klasycznego algorytmu genetycznego na proces minimalizacji testowej funkcji sferycznej Hypersphere. Poniżej mamy szczegółową analizę z podziałem na badane scenariusze.

---

### 1. Scenariusz: Wpływ wielkości populacji
Badanie pokazuje klasyczny przypadek kompromisu pomiędzy czasem obliczeń a zdolnością eksploracji przestrzeni rozwiązań:

* **Mała populacja (10 osobników):** Algorytm zadziałał najszybciej (**0.046 s**), jednak wynikiem tego jest przedwczesna zbieżność. Bardzo niska średnia wartość populacji (0.16) wskazuje na szybką utratę różnorodności genetycznej, co spowodowało utknięcie w 'prawie' optymalnym rozwiązaniu (**0.000016**).
* **Średnia populacja (50 osobników):** Stanowi idealny złoty środek. Osiągnęła optimum (**0.000000**) przy dobrym czasie (**0.228 s**).
* **Duża populacja (200 osobników):** Również znalazła idealne optimum, jednak czas obliczeń wzrósł ponad trzykrotnie (**0.768 s**). Wysoka wartość średnia populacji (5.35) pokazuje, że duża grupa osobników nieustannie i szeroko eksplorowała przestrzeń, nawet gdy najlepszy osobnik odnalazł już minimum.

<br>

### 2. Scenariusz: Metody selekcji
Wybór metody reprodukcji miał drastyczny wpływ na sukces optymalizacji:

* **Selekcja Turniejowa:** Okazała się bezkonkurencyjna, osiągając idealne optimum (**0.000000**). Zapewniła doskonały balans między promowaniem dobrych genów a utrzymaniem różnorodności.
* **Selekcja Ruletkowa:** Dała dobry, lecz minimalnie niedokładny wynik (**0.000022**). Mechanizm koła ruletki dał zbyt duże szanse na przetrwanie słabszym osobnikom.
* **Selekcja Najlepszych (Best):** Zakończyła się całkowitą porażką (wynik **0.108947**). Pozbycie się wszystkich słabych osobników na wczesnym etapie doprowadziło do ujednolicenia populacji, zatrzymując proces ewolucji.

<br>

### 3. Scenariusz: Skalowalność problemu
Eksperyment obrazuje wykładniczy wzrost trudności problemu wraz ze wzrostem liczby zmiennych decyzyjnych (N):

* Dla **N=2** problem był prosty – algorytm znalazł idealne zero (0.000000) w bardzo krótkim czasie.
* Dla **N=10** czas obliczeń wzrósł trzykrotnie (**0.537 s**), a wynik minimalnie stracił swoją precyzję (**0.000001**), choć nadal można uznać go za ogromny sukces.
* Dla **N=27** przestrzeń poszukiwań stała się tak ogromna, że liczba epok okazała się niewystarczająca. Czas wykonania był najdłuższy ze wszystkich testów minimalizacji (**1.24 s**), a wynik (**0.111854**) wskazuje, że algorytm nadal był w fazie schodzenia do minimum globalnego.

<br>

### 4. Scenariusz: Znaczenie Elitaryzmu
Zgodnie z teorią, ochrona najlepszego materiału genetycznego jest kluczowa dla stabilności. 

W teście `elita_brak` algorytm zgubił idealne rozwiązanie i zakończył z wynikiem **0.000001**, co jest dowodem na to, że w którejś z ostatnich epok operatory mutacji lub krzyżowania zniszczyły najlepszego osobnika. 

Wprowadzenie elity (`elita_2` oraz `elita_5`) całkowicie wyeliminowało ten problem, zachowując najlepszy wynik (**0.000000**). Warto zauważyć, że przy `elita=5` średnia populacji drastycznie spadła (0.82) – zbyt duża elita sprawia, że populacja staje się w "kopią liderów" i traci zdolność do szerokich poszukiwań.

<br>

### 5. Scenariusz: Metody Krzyżowania
W trakcie testów metod krzyżowania zaobserwować można anomalię. O ile krzyżowanie jednorodne (`uniform`) zadziałało perfekcyjnie (**0.000000**), o tyle metody `one_point`, `two_point` oraz `grainy` zakończyły działanie z wynikiem dokładnie **125.000000**.

**Błąd eksperymentu:** Wartość 125.0 nie jest przypadkowa. Dla funkcji Hypersphere o 5 wymiarach, maksimum na granicach dziedziny w przedziale od -5 do 5 wynosi dokładnie: 

`(-5)² * 5 = 125`
    
Dla tych trzech testów algorytm był optymalizowany do maksimum. Przyczyną tej anomalii jest zwykły błąd - pozostawienie parametru "Typ Optymalizacji" na wartości "Max" zamiast "Min" podczas konfiguracji. Ta sytuacja pokazuje, jak wrażliwe na warunki początkowe są eksperymenty optymalizacyjne.

---

### Podsumowanie Ostateczne
Najbardziej uniwersalną i stabilną konfiguracją okazał się **klasyczny algorytm genetyczny o populacji 50 osobników, wykorzystujący selekcję turniejową, krzyżowanie jednorodne (uniform) oraz strategię elitarną chroniącą 2 osobniki**. Taka parametryzacja gwarantuje optymalny balans pomiędzy czasem obliczeń a sukcesem eksploracji.
